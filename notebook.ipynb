# Book Recommender System

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

import warnings
warnings.filterwarnings("ignore")
```

```python
books = pd.read_csv('books.csv')
ratings = pd.read_csv('ratings.csv')
users = pd.read_csv('users.csv')
```

```python
books = books.drop(columns=['Image-URL-S', 'Image-URL-L'])
```

```python
books.head()
```

```python
# Lets remane columns name
books.rename(columns={"ISBN":"isbn",
                     "Book-Title":'title',
                      'Book-Author':'author',
                     "Year-Of-Publication":'year',
                     "Publisher":"publisher",
                     "Image-URL-L":"image_url"},inplace=True)
```

```python
ratings.head()
```

```python
ratings.shape
```

```python
# Lets remane columns name
ratings.rename(columns={"User-ID":'user_id',
                      'Book-Rating':'rating'},inplace=True)
```

```python
users_with_mt200_reviews = ratings['user_id'].value_counts() > 200
```

```python
users_with_mt200_reviews[users_with_mt200_reviews].shape
```

```python
users_id_with_mt200_reviews = users_with_mt200_reviews[users_with_mt200_reviews].index
```

```python
users_id_with_mt200_reviews
```

```python
ratings = ratings[ratings['user_id'].isin(users_id_with_mt200_reviews)]
```

```python
ratings.shape
```

```python
ratings.rename(columns={"ISBN":"isbn"},inplace=True)
```

```python
ratings.head()
```

```python
ratings_with_books = ratings.merge(books, on='isbn')
```

```python
ratings_with_books.head()
```

```python
number_rating = ratings_with_books.groupby('title')['rating'].count().reset_index()
```

```python
number_rating
```

```python
number_rating.rename(columns={'rating':'num_of_rating'},inplace=True)
```

```python
final_rating =ratings_with_books.merge(number_rating, on='title')
```

```python
print(final_rating.shape)
final_rating.head()
```

```python
final_rating = final_rating[final_rating['num_of_rating'] >= 50]
```

```python
final_rating.shape
```

```python
final_rating.drop_duplicates(['user_id','title'], inplace=True)
```

```python
final_rating.shape
```

```python
book_pivot = final_rating.pivot_table(columns='user_id', index='title', values='rating')
```

```python
book_pivot.head(5)
```

```python
book_pivot.shape
```

```python
from sklearn.preprocessing import StandardScaler
scaler = StandardScaler()
book_pivot_scaled = pd.DataFrame(
    scaler.fit_transform(book_pivot.fillna(0)), 
    index=book_pivot.index, 
    columns=book_pivot.columns
)
```

```python
book_pivot.head()
```

```python
from scipy.sparse import csr_matrix

book_sparse = csr_matrix(book_pivot.fillna(0).values)
```

# Training Model

```python
from sklearn.neighbors import NearestNeighbors
model = NearestNeighbors(metric='cosine', algorithm='brute')
```

```python
model.fit(book_sparse)
```

```python
_, suggestion = model.kneighbors(book_sparse[100], n_neighbors=6)
```

```python
suggestion
```

```python
for i in range(len(suggestion[0])):
    print(book_pivot.index[suggestion[0][i]])
```

# Finding Poster Links

```python
book_name = []
for book_id in suggestion:
    book_name.append(book_pivot.index[book_id])
```

```python
book_name[0]
```

```python
ids_index = []
for name in book_name[0]:
    ids = np.where(final_rating['title'] == name)[0][0]
    ids_index.append(ids)
```

```python
for idx in ids_index:
    url = final_rating.iloc[idx]['Image-URL-M']
    print(url)
```

```python
book_names= list(book_pivot.index)
```

```python
import pickle
pickle.dump(model,open('model.pkl','wb'))
pickle.dump(book_names,open('book_names.pkl','wb'))
pickle.dump(final_rating,open('final_rating.pkl','wb'))
pickle.dump(book_pivot,open('book_pivot.pkl','wb'))
pickle.dump(book_sparse,open('book_sparse.pkl','wb'))
```

# Testing Model

```python
def recommend_book(book_name):
    book_id = np.where(book_pivot.index == book_name)[0][0]
    query = book_sparse[book_id]
    _, suggestion = model.kneighbors(query, n_neighbors=6)   
    recommended_indices = suggestion[0][1:]  
    
    for i in recommended_indices:
        print(book_pivot.index[i])
```

```python
book_name = "1984"
recommend_book(book_name)
```
